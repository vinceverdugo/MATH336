### 3.1  Use Python to define a data sequence t=seq(2015,2018, length=100), and then plot the following two functions on the same figure: 
#### y = sin(2π(t − 0.1)) and y = cos^2(2πt).

```python
import matplotlib.pyplot as plt
import numpy as np
#set 't' parameters for plot on (t,y) axis
t = np.linspace(2015, 2018, num=1000)
plt.figure(figsize=(6,6))
#plot both functions
plt.plot(t, np.sin(2*np.pi*(t-0.1)), 'b-', 
         label = f"y = sin(2π(t-0.1)))")
plt.plot(t, np.cos(2*np.pi*t)**2, 'r-',
        label = f" y = cos$^2$(2πt)")
#where to place legend
plt.legend(loc='best')
```

### 3.4 Use Python to solve the following linear equations:

#### −3x + 2y + z = 1

#### −2x − y + z = 2

#### 2x + y − 4z = 0

```python
from sympy import symbols, Eq, solve
x, y, z = symbols('x y z')
#create system of equations
eq1 = Eq(-3*x + 2*y + z, 1)
eq2 = Eq(-2*x - y + z, 2)
eq3 = Eq(2*x + y - 4*z, 0)
#use solve function from sympy to solve for x, y, and z
sol = solve((eq1, eq2, eq3), (x, y, z))

print(f"x = {sol[x]}, y = {sol[y]}, z = {sol[z]}")
```

### 3.5 Surface air temperature (SAT) is often defined as the temperature inside a white-painted louvered instrument container or box, known as a Stevenson screen located on a stand about 2 meters above the ground. The daily minimum temperature (Tmin) is the minimum temperature within 24 hours. The daily mean temperature (Tmean) is the average of Tmax and Tmin.

### 3.7 
#### a) Use Python to plot the the Cuyamaca January Tmin time series from 1951 to 2010 with a continuous curve.

```python
#collect January specific data and organize into array
janTminArray = np.array(cuyamacaData.iloc[768:1477, 5:6].iloc[::12]).T
#years array
years = np.arange(1951, 2011)
```

```python
#define linear model function from book
def linear_model(x):
    return lambda t: x[0]*t + x[1]
```

```python
#create linear model
tTrend = np.polyfit(years, janTminArray[0].astype(float), 1)
print(f"Slope of linear trend: {tTrend[0]:5.4f}, "
      f"y-intercept of linear trend: {tTrend[1]:5.2f}")
tminLine = linear_model(tTrend)
```

```python
#plot data
plt.figure(figsize=(12,9))
plt.plot(years, janTminArray[0].astype(float), 'k-o', mfc='none')
plt.plot(years, tminLine(years), 'r-',
        label=f"Trend = {tTrend[0]:.3f} per year")
plt.xlabel("Year", size=12)
plt.ylabel(f"Temperature ($\degree$F)", size=12)
plt.legend(loc = "best")
plt.title("January Tmin Values at Cuyamaca Station, 1951-2010")
```
